# Configuration model

Configuration model is a model to generate a random graph with given distribution of degree. 

Algorithm 

1. Determine a distribution of degree $\{p(0), p(1), p(2), \cdots \}$. 
2. Determine $k$ by following the probability $p(k)$ and generate the $n$ th degree vector $[k_1, 3, 4, 4]$

Please note that self-loop and double loop can be allowed.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
degree_sequence = [1, 2, 2, 3, 4, 4]
G = nx.configuration_model(degree_sequence)
nx.draw(G, with_labels=True)
```

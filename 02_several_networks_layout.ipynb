# Several networks and a layout of network

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 16
```

```python
# Generate a random graph (Erdos-Renyi model) 
G_rnd = nx.gnp_random_graph(10, 0.1)

## spring layout
pos = nx.spring_layout(G_rnd)
plt.title('Random Graph with spring layout')
nx.draw(G_rnd, pos, with_labels=True)
plt.show()

## circular layout
pos = nx.circular_layout(G_rnd)
plt.title('Random Graph with circular layout')
nx.draw(G_rnd, pos, with_labels=True)
plt.show()

## random layout
pos = nx.random_layout(G_rnd)
plt.title('Random Graph with random layout')
nx.draw(G_rnd, pos, with_labels=True)
plt.show()
```

```python
# Generate a cycle graph
G_cycle = nx.cycle_graph(10)
pos = nx.spring_layout(G_cycle)

plt.title('Cycle Graph')
nx.draw(G_cycle, pos, with_labels=True)
```

```python
# Generate a complete graph
G_K10 = nx.complete_graph(10)
pos = nx.spring_layout(G_K10)

plt.title('Complete Graph with K = 10')
nx.draw(G_K10, pos, with_labels=True)
```

```python
# Generate a star graph
G_star = nx.star_graph(9)
pos = nx.spring_layout(G_star)

plt.title('Star Graph')
nx.draw(G_star, pos, with_labels=True)
```

```python
# Generate a wheel graph
G_wheel = nx.wheel_graph(10)
pos = nx.spring_layout(G_wheel)

plt.title('Wheel Graph')
nx.draw(G_wheel, pos, with_labels=True)
```

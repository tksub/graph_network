# Degree and distribution of degree



```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
# Generate a simple graph
G = nx.Graph([(1, 2), (1, 3), (1, 4), (1, 5), (4, 5)])
print(nx.info(G))
nx.draw(G, with_labels=True)
```

```python
print(nx.degree_histogram(G))
degree_dist = [i/5 for i in nx.degree_histogram(G)]

plt.rcParams['font.size'] = 20
plt.bar(range(5), height=degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0, 1)
```

## Degree distribution for different kinds of graphs

```python
N = 10

# Random Graph
G_rnd = nx.gnp_random_graph(N, 0.1)
pos = nx.circular_layout(G_rnd)

# Info.
print(nx.info(G_rnd))
print('density = ', nx.density(G_rnd), 'average clustering coefficient = ', nx.average_clustering(G_rnd))

# Draw graph
plt.subplot(1,2,1)
plt.title('Random Graph')
nx.draw(G_rnd, pos, with_labels=True)

# Degree distribution
plt.subplot(1,2,2)
degree_dist = [ind/N for ind in nx.degree_histogram(G_rnd)]
plt.plot(degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0, 1)
```

```python
N = 10

# Cycle Graph
G_cycle= nx.cycle_graph(N)
pos = nx.circular_layout(G_cycle)

# Info.
print(nx.info(G_cycle))
print('density = ', nx.density(G_rnd), 'average clustering coefficient = ', nx.average_clustering(G_rnd))

# Draw graph
plt.subplot(1,2,1)
plt.title('Cycle Graph')
nx.draw(G_cycle, pos, with_labels=True)

# Degree distribution
plt.subplot(1,2,2)
degree_dist = [ind/N for ind in nx.degree_histogram(G_cycle)]
plt.plot(degree_dist)
plt.xlabel('$k$')
plt.ylabel('$p(k)$')
plt.ylim(0, 1)
```

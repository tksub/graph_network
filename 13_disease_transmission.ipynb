# SIR simulation with Gillespie algorithm

Gillespie algorithm for probablic disease transmission simulation 

S: safe
I: infected with probability $\beta$
R: Revive with probability $\gamma$

1. Generate a network by `networkx`
2. Simulate the disease transmission by `EoM` + Gillespie algorithm 
3. Show results


NDlib is similar package as EoN

pip install EoN

````python
import EoN
import matplotlib.pyplot as plt
import networkx as nx
````

````python
N = 10**5
G = nx.barabasi_albert_graph(N, 5)

# parameters for simulation
gamma = 1.0
tau = 0.3
rho = 0.005 # initial infection

# SIR simulation with Gillespie algorithm 
t, S, I, R = EoN.Gillespie_SIR(G, tau, gamma, rho=rho)

plt.plot(t, S, color='green', label='S')
plt.plot(t, I, color='red', label='I')
plt.plot(t, R, color='blue', label='R')
plt.xlabel('t')
plt.legend()
plt.show()
````

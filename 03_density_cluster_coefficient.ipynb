# Density and Clustering coefficient of graph

## Density
Density $\rho$ is given by 
$$
\rho = \frac{m}{n(n-1)/2} = \frac{2m}{n(n-1)},
$$
where $m$ and $n$ represent the number of edges and pairs between nodes, respectively. 

## Clustering coefficient 
A clustering coefficient of $i$ th node, $C_i$ is a measure how many neighbor nodes are connected.

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
# Generate a network
G = nx.Graph([(1, 2), (1, 3), (1, 4), (1, 5), (3, 4), (4, 5), (4, 6), (5, 6)])
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
print('Averaged clustering coefficient = ', nx.average_clustering(G))
```

```python
# Random Graph
G_rnd = nx.gnp_random_graph(10, 0.1)
pos = nx.circular_layout(G_rnd)

plt.title('Random Graph')
nx.draw(G_rnd, pos, with_labels=True)
print('density = ', nx.density(G_rnd), 'average clustering coefficient = ', nx.average_clustering(G_rnd))
```

```python
# Cycle Graph
G_cycle = nx.cycle_graph(10)
pos = nx.spring_layout(G_cycle)

plt.title('Cycle Graph')
nx.draw(G_cycle, pos, with_labels=True)
print('density = ', nx.density(G_cycle), 'average clustering coefficient = ', nx.average_clustering(G_cycle))
```

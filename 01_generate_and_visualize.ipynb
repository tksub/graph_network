# Generate and visualize a simple network

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
# Generate a graph
G = nx.Graph([(1, 2), (2,3), (3,1)])
nx.draw(G, with_labels=True)
```

```python
# Add a node
G.add_node(4)
nx.draw(G, with_labels=True)
```

```python
# Add the edge between the nodes 1 and 4
G.add_edge(1,4)
nx.draw(G, with_labels=True)
```

```python
# Add multiple nodes
n = [5, 6, 7]
G.add_nodes_from(n)

# Add multiple edges
e = [(5,6), (3,7), (4,6)]
G.add_edges_from(e)

nx.draw(G, with_labels=True)
```

```python
# Remove a node
G.remove_node(4)
nx.draw(G, with_labels=True)
```

```python
# Remove multiple edges
G.remove_edges_from([(1, 2), (2, 3)])

nx.draw(G, with_labels=True)
```

# Centrality-1

The motivation is to figure out which node is a central one in a network.


In this code, the following two centralities will be focused. 

- Degree Centrality 
- Closeness Centrality


## Degree centrality

The number of edges of a specific node is the degree centrality. In the `networkx`, outputted degree centrality is normalized by $N-1$. Here $N-1$ is the maximum nodes that a node can be connected. 

## Closeness centrality
Closeness centrality is defined as the inverse of average distance of all nodes from a specific node. When neighboring centrality $l_i$ is defined as 

$$ 
 l_i = \frac{\sum_{j}d_{ij}}{N-1}
$$

so, the closeness centrality is defined as $l_i^{-1}$. 

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
# Generate an example
G = nx.Graph([(1, 2), (1, 3), (1, 4), (1, 5), (3, 4), (4, 5), (4, 6), (5, 6)])
nx.draw(G, with_labels=True)
```

```python
print('Degree Centrality = ', nx.degree_centrality(G))
print('Closeness Centrality = ', nx.closeness_centrality(G))
```
